```python
# as we can see in the previous table, the majority of the null columns are plenty of them.
# so that I consider adecuate to drop down the columns with 40% or more nulls:

threshold = 40
null_cols = null_cols[null_cols > 0] / len(df_merge) * 100
null_filter = null_cols > threshold
drop_cols = list(null_cols[null_filter].index)
drop_cols
```

```python
df_merge_2 = df_merge.drop(drop_cols, axis=1)
len(df_merge_2.columns)
```
